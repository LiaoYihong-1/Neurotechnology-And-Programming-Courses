Пусть $f(x) = \sin(x) * x$. Постройте график $f(x)$ и $\frac{df(x)}{dx}$, для расчета производных воспользуйтесь методом `backward()` и matplotlib для построения графиков.

```python
import torch
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

x = torch.arange(-20, 20, 0.2, requires_grad=True)
y = x.sin() * x
plt.plot(x.detach().numpy(), y.detach().numpy(), label='1')
y.sum().backward()
plt.plot(x.detach().numpy(), x.grad.detach().numpy(), label='function')
plt.show()
```
